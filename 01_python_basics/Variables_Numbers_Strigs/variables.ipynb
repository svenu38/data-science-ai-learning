### ðŸ”¹ **Variable Naming Rules**
- Must start with a letter or underscore (`_`)
- Cannot start with a number  
- Cannot contain spaces  
- Cannot use special characters (`@`, `#`, `-`, `%`, etc.)  
- Case-sensitive (`name` and `Name` are different)

## âž¤ **Basic Arithmetic Operators**

| Operator | Meaning | Example | Result |
|---------|---------|---------|--------|
| `+` | Addition | `a + b` | Adds values |
| `-` | Subtraction | `a - b` | Difference |
| `*` | Multiplication | `a * b` | Product |
| `/` | Division | `a / b` | Returns float result |
| `//` | Floor Division | `a // b` | Removes decimals |
| `%` | Modulus | `a % b` | Remainder |
| `**` | Exponent (power) | `a ** b` | a raised to b |

# ðŸŽµ String Indexing in Python (Example: "Shankar Mahadevan")

A **string** in Python is a sequence of characters.  
Each character has a **position** called an *index*.

Python supports two types of indexing:

- **Positive Indexing** â†’ counts from the front (starting at 0)  
- **Negative Indexing** â†’ counts from the end (starting at -1)

Let's use the name **"Shankar Mahadevan"** to understand indexing.

---

## ðŸ”¹ Positive Indexing (0, 1, 2, ...)



```python
full_name
```

```python
full_name[0]
```

```python
len(full_name)
```

```python
print(full_name[0])   # S
print(full_name[1])   # h
print(full_name[7])   # space
print(full_name[8])   # M
print(full_name[16])  # n
```

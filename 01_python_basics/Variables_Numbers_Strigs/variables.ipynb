### üîπ **Variable Naming Rules**
- Must start with a letter or underscore (`_`)
- Cannot start with a number  
- Cannot contain spaces  
- Cannot use special characters (`@`, `#`, `-`, `%`, etc.)  
- Case-sensitive (`name` and `Name` are different)
